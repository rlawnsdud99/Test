```python
import tensorflow as tf
import numpy as np
x_data = np.array([[0, 0], [0, 1], [1, 0], [1, 1]], dtype=np.float)
y_data = np.array([[0], [1], [1], [0]], dtype=np.float)
#XOR은 x,y가 같을경우 0을, 다를경우 1을 출력하므로 그에 해당하는 정답 데이터 를 만든다
#X가 입력, Y가 출력이다.
#djWdklsfhjkashdkfjhaweilk

#Keras를 이용해 간편히 모델을 만든다. 
model = tf.keras.models.Sequential([
    tf.keras.layers.Dense(16, input_shape=(2,), activation='sigmoid'),
    tf.keras.layers.Dense(1, activation='sigmoid')
])
#Dense 층을 2개 이용하여 네트워크를 구성하고. activation은 sigmoid로 이용한다.

#옵티마이저는 SGD를 이용한다. 
SGD = tf.keras.optimizers.SGD(lr=1.5)

#모델, loss func 평가 방법 등등을 모델에 컴파일한다.
model.compile(optimizer=SGD, loss='mse', metrics=['accuracy'])

#모델을 500회 학습시킨다.
model.fit(x_data, y_data, epochs=500)

#모델을 평가하고 정답지에 대한 출력을 실험해본다.
model.evaluate(x_data, y_data)
predict = model.predict(x_data)
print(predict)
```

```python
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np

#minist dataset을 불러온다
mnist = tf.keras.datasets.mnist

#입출력 구분, 학습용과 테스트용을 구분한다
(x_train, y_train), (x_test, y_test) = mnist.load_data() 

#출력데이터를 onehot형식으로 바꾼다(일렬 데이터로 변환).
#또한 입력은 0~1 사이의 값으로 정규화한다
y_train = tf.keras.utils.to_categorical(y_train)
y_test = tf.keras.utils.to_categorical(y_test) 
x_train, x_test = x_train/255.0, x_test/255.0
print(x_train.shape, y_train.shape, x_test.shape, y_test.shape) 

#Sequential keras 모델을 만든다.
model = tf.keras.models.Sequential([
    #입력되는 이미지가 28*28의 형태이므로, Flatten을 통해 1차원 배열로 바꿔준다
    tf.keras.layers.Flatten(input_shape=(28,28)), 
    #DNN 네트워크를 만든다. Drop out을 추가하여, 과적합을 막는다.
    tf.keras.layers.Dense(128, activation='relu'), 
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Dense(128, activation='relu'), 
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Dense(10, activation='softmax') 
])
#모델을 요약한다
model.summary()

#optimizer, loss function, 평가방식을 정하여 모델을 컴파일한다
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['categorical_accuracy'])

#모델을 학습한다
model.fit(x_train, y_train, epochs=20, validation_data=(x_test, y_test), verbose=2) 

#모델을 평가한다
model.evaluate(x_test, y_test, verbose=2)

#랜덤이미지에 대해 적용해본다
random_image = x_test[np.random.choice(len(x_test))] 
test_image = np.expand_dims(random_image, axis = 0) 
predict = model.predict(test_image)
print('predict ',predict, np.argmax(predict, axis=1)) 
plt.imshow(test_image[0], cmap='Greys')
plt.show()
```

```python
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
#minist dataset을 불러온다
mnist = tf.keras.datasets.mnist

#입출력 구분, 학습용과 테스트용을 구분한다
(x_train, y_train), (x_test, y_test) = mnist.load_data() 

#출력데이터를 onehot형식으로 바꾼다(일렬 데이터로 변환).
#또한 입력은 0~1 사이의 값으로 정규화한다
y_train = tf.keras.utils.to_categorical(y_train)
y_test = tf.keras.utils.to_categorical(y_test) 
x_train, x_test = x_train/255.0, x_test/255.0

print(x_train.shape, y_train.shape, x_test.shape, y_test.shape) 

#Sequential keras 모델을 만든다.
model = tf.keras.models.Sequential([
    #CNN부분 2차원 이미지를 3차원 배열로 넣고, Conv filter를 이용해, 특징을 잡고, 
    #MaxPool을 통해 더 특징적인 부분을 볼수 있다.(ex)얼굴 전체를 보기보다, 코 눈 입을 볼수있게된다 )
    
    tf.keras.layers.Reshape((28,28,1),input_shape=(28,28)), 
    tf.keras.layers.Conv2D(filters=32,kernel_size=3, padding='same', activation='relu'),
    tf.keras.layers.MaxPool2D(pool_size=(2,2)),

    tf.keras.layers.Conv2D(filters=64,kernel_size=3, padding='same', activation='relu'),
    tf.keras.layers.MaxPool2D(pool_size=(2,2)),
 
    tf.keras.layers.Conv2D(filters=128,kernel_size=3, padding='same', activation='relu'),
    tf.keras.layers.MaxPool2D(pool_size=(2,2)),
    
    #DNN부분.뽑아진 특징들을 1차원으로 만들어 간단한 .DNN에 넣고, 끝에서 출력한다
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(10, activation='softmax') 
])

#모델을 요약한다
model.summary()

#optimizer, loss function, 평가방식을 정하여 모델을 컴파일한다
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['categorical_accuracy'])

#학습한다
model.fit(x_train, y_train, epochs=5, validation_data=(x_test, y_test), verbose=2) 
#학습된 모델을 평가한다
model.evaluate(x_test, y_test, verbose=2)

#학습된 모델과 가중치를 저장한다.
model.save('mnist_cnn.h5')
model.save_weights('mnist_cnn_weight.h5')
print("Model and Weights succesfully Saved-mnist_cnn.h5")

#랜덤 이미지에대해 적용해본다
random_image = x_test[np.random.choice(len(x_test))] 
test_image = np.expand_dims(random_image, axis = 0) 

predict = model.predict(test_image)
print('predict ',predict, np.argmax(predict, axis=1)) 

plt.imshow(test_image[0], cmap='Greys')
plt.show()
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Conv2D, Flatten, Dropout, MaxPooling2D
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import os
import numpy as np
import matplotlib.pyplot as plt

#데이터셋을 불러오고 압축을 푼다
_URL = 'https://storage.googleapis.com/mledu-datasets/cats_and_dogs_filtered.zip' 
path_to_zip = tf.keras.utils.get_file('cats_and_dogs.zip', origin=_URL, extract=True) 
PATH = os.path.join(os.path.dirname(path_to_zip), 'cats_and_dogs_filtered')

#학습용과 테스트용으로 나눠가져온다
train_dir = os.path.join(PATH, 'train')
validation_dir = os.path.join(PATH, 'validation')

#다시 개와 고양이로 나눠 저장한다
train_cats_dir = os.path.join(train_dir, 'cats')  
train_dogs_dir = os.path.join(train_dir, 'dogs')

validation_cats_dir = os.path.join(validation_dir, 'cats')  
validation_dogs_dir = os.path.join(validation_dir, 'dogs') 

#각 학습,테스트,개,고양이 데이터이미지의 수를 확인한다
num_cats_tr = len(os.listdir(train_cats_dir))
num_dogs_tr = len(os.listdir(train_dogs_dir))
num_cats_val = len(os.listdir(validation_cats_dir))
num_dogs_val = len(os.listdir(validation_dogs_dir))
total_train = num_cats_tr + num_dogs_tr 
total_val = num_cats_val + num_dogs_val
print('total training cat images:', num_cats_tr)
print('total training dog images:', num_dogs_tr)
print('total validation cat images:', num_cats_val)
print('total validation dog images:', num_dogs_val)
print("--")
print("Total training images:", total_train)
print("Total validation images:", total_val)

#학습용 parameter를 설정한다.
batch_size = 32 
epochs = 15 
IMG_HEIGHT = 150 
IMG_WIDTH = 150

train_image_generator = ImageDataGenerator(rescale=1./255) 
validation_image_generator = ImageDataGenerator(rescale=1./255)

image_gen_train = ImageDataGenerator(rescale=1./255, 
                                     rotation_range=45, 
                                     width_shift_range=.15, 
                                     height_shift_range=.15, 
                                     horizontal_flip=True, 
                                     zoom_range=0.5
)

train_data_gen = image_gen_train.flow_from_directory(batch_size=batch_size,
                                                     directory=train_dir, shuffle=True,
                                                     target_size=(IMG_HEIGHT, IMG_WIDTH), 
                                                     class_mode='binary')


val_data_gen = validation_image_generator.flow_from_directory(batch_size=batch_size,
                                                              directory=validation_dir,
                                                              target_size=(IMG_HEIGHT, IMG_WIDTH), 
                                                              class_mode='binary')

augmented_images = [train_data_gen[0][0][0] for i in range(5)] 
plotImages(augmented_images)

"""model = Sequential([
    Conv2D(16, 3, padding='same', activation='relu', 
           input_shape=(IMG_HEIGHT, IMG_WIDTH ,3)),
    MaxPooling2D(),
    Dropout(0.2),
    Conv2D(32, 3, padding='same', activation='relu'), 
    MaxPooling2D(),
    Dropout(0.2),
    Conv2D(64, 3, padding='same', 
           activation='relu'), 
    MaxPooling2D(),
    Dropout(0.2),
    Flatten(),
    Dense(512, activation='relu'), 
    Dense(1, activation='sigmoid')
])
"""
vgg = tf.keras.applications.VGG19(include_top=False, 
                                  weights='imagenet', 
                                  input_shape=(IMG_HEIGHT, IMG_WIDTH ,3))

model_vgg = Sequential([
    vgg,
    tf.keras.layers.GlobalAveragePooling2D(), 
    Dense(1, activation='sigmoid')
])

"""model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.summary()
"""

adam = tf.keras.optimizers.Adam(lr=0.0001) 
model_vgg.compile(optimizer=adam,loss='binary_crossentropy', metrics=['accuracy'])
model_vgg.summary()

history = model_vgg.fit_generator( 
    train_data_gen,
    steps_per_epoch=total_train // batch_size, 
    epochs=epochs,
    validation_data=val_data_gen, 
    validation_steps=total_val // batch_size, 
    verbose=2
)

acc = history.history['acc']
val_acc = history.history['val_acc'] 

loss = history.history['loss']
val_loss = history.history['val_loss'] 

epochs_range = range(epochs)

plt.figure(figsize=(8, 8))
plt.subplot(1, 2, 1)
plt.plot(epochs_range, acc, label='Training Accuracy')
plt.plot(epochs_range, val_acc, label='Validation Accuracy')
plt.legend(loc='lower right')
plt.title('Training and Validation Accuracy')

plt.subplot(1, 2, 2)
plt.plot(epochs_range, loss, label='Training Loss')
plt.plot(epochs_range, val_loss, label='Validation Loss')
plt.legend(loc='upper right')
plt.title('Training and Validation Loss')
plt.show()
```
